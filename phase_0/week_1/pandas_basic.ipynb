## Columns Manipulation

- create new columns
- rename column
- drop column

```python
# membuat salinanb
df_new = df.copy()
```

```python
df_new.head()
```

```python
# create new column
df_new['selisih'] = np.abs(df_new['pts'] - df_new['opp_pts'])
```

```python
df_new.head()
```

rename column

```python
df_new = df_new.rename(columns={'game_location': 'loc',
                        'game_result': 'result'})
```

```python
df_new
```

drop column

```python
df_new.drop(['elo_i', 'elo_n'], axis=1, inplace=True)
```

```python
df_new
```

drop rows
- drop year = 1947 & team spesifik - query
- index kita salin jadikan list
- list sebagai input .drop, axis = 0

```python
# creating new column based on existing information using loop

# siapkan list kosong
# buat program loop -> selisih pts dengan opp_pts per baris
# simpan hasil per baris ke list kosong
# convert list yang sudah terisi menjadi kolom

hasil = []

# df_new.iloc[:, 10] - kolom pts
# df_new.iloc[:, 14] - kolom opp_pts

for i in range(len(df_new)):
    selisih = df_new.iloc[i, 10] - df_new.iloc[i, 14]
    hasil.append(selisih)

df_new['menggunakan_loop'] = hasil
```
